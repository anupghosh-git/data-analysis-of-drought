```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

olpe_216 = pd.read_csv('produkt_rr_stunde_20041001_20191231_00216.txt', sep=";")
olpe_5468 = pd.read_csv('produkt_rr_stunde_20060701_20191231_05468.txt', sep=";")
olpe_2947 = pd.read_csv('produkt_rr_stunde_20061001_20191231_02947.txt', sep=";")

hsk_1300 = pd.read_csv('produkt_rr_stunde_20040601_20191231_01300.txt',sep=";")
hsk_2483 = pd.read_csv('produkt_rr_stunde_19951012_20191231_02483.txt',sep=";")
hsk_3215 = pd.read_csv('produkt_rr_stunde_20070601_20191231_03215.txt',sep=";")
hsk_4488 = pd.read_csv('produkt_rr_stunde_20060801_20191231_04488.txt',sep=";")
hsk_6264 = pd.read_csv('produkt_rr_stunde_20040601_20191231_06264.txt',sep=";")
hsk_7330 = pd.read_csv('produkt_rr_stunde_20051001_20191231_07330.txt',sep=";")
```

```python
olpe_216['MESS_DATUM'] = pd.to_datetime(olpe_216['MESS_DATUM'], format='%Y%m%d%H')
olpe_216 = olpe_216[olpe_216.MESS_DATUM.between('2017-04-16','2017-08-16')]

olpe_5468['MESS_DATUM'] = pd.to_datetime(olpe_5468['MESS_DATUM'], format='%Y%m%d%H')
olpe_5468 = olpe_5468[olpe_5468.MESS_DATUM.between('2017-04-16','2017-08-16')]

olpe_2947['MESS_DATUM'] = pd.to_datetime(olpe_2947['MESS_DATUM'], format='%Y%m%d%H')
olpe_2947 = olpe_2947[olpe_2947.MESS_DATUM.between('2017-04-16','2017-08-16')]
```

```python
hsk_1300['MESS_DATUM'] = pd.to_datetime(hsk_1300['MESS_DATUM'], format='%Y%m%d%H')
hsk_1300 = hsk_1300[hsk_1300.MESS_DATUM.between('2017-04-16','2017-08-16')]

hsk_2483['MESS_DATUM'] = pd.to_datetime(hsk_2483['MESS_DATUM'], format='%Y%m%d%H')
hsk_2483 = hsk_2483[hsk_2483.MESS_DATUM.between('2017-04-16','2017-08-16')]

hsk_3215['MESS_DATUM'] = pd.to_datetime(hsk_3215['MESS_DATUM'], format='%Y%m%d%H')
hsk_3215 = hsk_3215[hsk_3215.MESS_DATUM.between('2017-04-16','2017-08-16')]

hsk_4488['MESS_DATUM'] = pd.to_datetime(hsk_4488['MESS_DATUM'], format='%Y%m%d%H')
hsk_4488 = hsk_4488[hsk_4488.MESS_DATUM.between('2017-04-16','2017-08-16')]

hsk_6264['MESS_DATUM'] = pd.to_datetime(hsk_6264['MESS_DATUM'], format='%Y%m%d%H')
hsk_6264 = hsk_6264[hsk_6264.MESS_DATUM.between('2017-04-16','2017-08-16')]

hsk_7330['MESS_DATUM'] = pd.to_datetime(hsk_7330['MESS_DATUM'], format='%Y%m%d%H')
hsk_7330 = hsk_7330[hsk_7330.MESS_DATUM.between('2017-04-16','2017-08-16')]
```

```python
df = pd.DataFrame(olpe_216, columns = ['MESS_DATUM'])
df.index = np.arange(1,len(df)+1)
df
```

```python
df['R_216_olpe'] = olpe_216['  R1'].values
df['R_216_olpe_cum'] = df['R_216_olpe'].cumsum()
df['R_5468_olpe'] = olpe_5468['  R1'].values
df['R_5468_olpe_cum'] = df['R_5468_olpe'].cumsum()
df["R_2947_olpe"] = olpe_2947['  R1'].values
df['R_2947_olpe_cum'] = df['R_2947_olpe'].cumsum()
df['R_mean_olpe'] = (df['R_216_olpe']+df['R_5468_olpe']+df['R_2947_olpe'])/3.0
df['R_mean_olpe_cum'] = df['R_mean_olpe'].cumsum()
df
```

```python
df['R_1300_hsk'] = hsk_1300['  R1'].values
df['R_1300_hsk_cum'] = df['R_1300_hsk'].cumsum()
df['R_2483_hsk'] = hsk_2483['  R1'].values
df['R_2483_hsk_cum'] = df['R_2483_hsk'].cumsum()
df["R_3215_hsk"] = hsk_3215['  R1'].values
df['R_3215_hsk_cum'] = df['R_3215_hsk'].cumsum()
df['R_4488_hsk'] = hsk_4488['  R1'].values
df['R_4488_hsk_cum'] = df['R_4488_hsk'].cumsum()
df['R_6264_hsk'] = hsk_6264['  R1'].values
df['R_6264_hsk_cum'] = df['R_6264_hsk'].cumsum()
df["R_7330_hsk"] = hsk_7330['  R1'].values
df['R_7330_hsk_cum'] = df['R_7330_hsk'].cumsum()
df['R_mean_hsk'] = (df['R_1300_hsk']+df['R_2483_hsk']+df['R_3215_hsk']+df['R_4488_hsk']+df['R_6264_hsk']+df['R_7330_hsk'])/6.0
df['R_mean_hsk_cum'] = df['R_mean_hsk'].cumsum()
df
```

```python
from datetime import datetime
from matplotlib.dates import DateFormatter
myFmt = DateFormatter("%d-%m-%y")
#axes1[0].xaxis.set_major_formatter(myFmt)

x1 = df.MESS_DATUM

# Defining custom 'xlim' and 'ylim' values.
custom_ylim = (0, 500)

fig, axes1 = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))

axes1[0].plot(x1, df.R_216_olpe_cum)
axes1[0].set_title('Olpe_Station_ID-00216')
axes1[0].set_ylabel('Cumulative Precipitation (mm)')

axes1[1].plot(x1, df.R_2947_olpe_cum)
axes1[1].set_title('Olpe_Station_ID-02947')
axes1[1].set_ylabel('Cumulative Precipitation (mm)')

axes1[2].plot(x1, df.R_5468_olpe_cum)
axes1[2].set_title('Olpe_Station_ID-05468')
axes1[2].set_ylabel('Cumulative Precipitation (mm)')

# Setting the values for all axes.
plt.setp(axes1, ylim=custom_ylim)
# plt.savefig('Olpe_line_graph.png')
fig.autofmt_xdate()

fig, axes2 = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))
axes2[0].plot(x1, df.R_1300_hsk_cum, 'orange')
axes2[0].set_title('HSK_Station_ID-01300')
axes2[0].set_ylabel('Cumulative Precipitation (mm)')

axes2[1].plot(x1, df.R_2483_hsk_cum, 'orange')
axes2[1].set_title('HSK_Station_ID-02483')
axes2[1].set_ylabel('Cumulative Precipitation (mm)')

axes2[2].plot(x1, df.R_3215_hsk_cum, 'orange')
axes2[2].set_title('HSK_Station_ID-03215')
axes2[2].set_ylabel('Cumulative Precipitation (mm)')

# Setting the values for all axes.
plt.setp(axes2, ylim=custom_ylim)

fig.autofmt_xdate()

fig, axes3 = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))
axes3[0].plot(x1, df.R_4488_hsk_cum, 'orange')
axes3[0].set_title('HSK_Station_ID-04488')
axes3[0].set_ylabel('Cumulative Precipitation (mm)')

axes3[1].plot(x1, df.R_6264_hsk_cum, 'orange')
axes3[1].set_title('HSK_Station_ID-06264')
axes3[1].set_ylabel('Cumulative Precipitation (mm)')

axes3[2].plot(x1, df.R_7330_hsk_cum, 'orange')
axes3[2].set_title('HSK_Station_ID-07330')
axes3[2].set_ylabel('Cumulative Precipitation (mm)')

# Setting the values for all axes.
plt.setp(axes3, ylim=custom_ylim)

fig.autofmt_xdate()

```

```python
df_day = df.groupby(pd.Grouper(key='MESS_DATUM',freq='D')).sum()
# df_day['DATE'] = df_day.index
# df_day.index = np.arange(1,len(df_day)+1)
df_day['DATE'] = pd.to_datetime(df_day.index)
```

```python
df_day
```

```python
df_day_april = (df_day['DATE'] >= '2017-04-16') & (df_day['DATE'] <= '2017-05-15')
df_day_may = (df_day['DATE'] >= '2017-05-16') & (df_day['DATE'] <= '2017-06-15')
df_day_june = (df_day['DATE'] >= '2017-06-16') & (df_day['DATE'] <= '2017-07-15')
df_day_july = (df_day['DATE'] >= '2017-07-16') & (df_day['DATE'] <= '2017-08-16')

df_day_april = df_day.loc[df_day_april]
df_day_may = df_day.loc[df_day_may]
df_day_june = df_day.loc[df_day_june]
df_day_july = df_day.loc[df_day_july]
```

```python
import matplotlib.dates as mdates

def line_format(label):
    """
    Convert time label to the format of pandas line plot
    """
    month = label.month_name()[:3]
#     if month == 'Jan':
#         month += f'\n{label.year}'
    month+=f' {label.day}'
    return month

custom_ylim = (0, 40)

labels = df_day_april.DATE
x = np.arange(len(labels))
width = 0.5
fig, ax = plt.subplots(figsize=(20,7))
ax.bar(x-width/2, df_day_april.R_mean_olpe, width, label = 'Olpe')
ax.bar(x+width/2, df_day_april.R_mean_hsk, width, label = 'HSK')

ax.set_xticks(x)
ax.set_xticklabels(map(line_format, df_day_april.index))

ax.set_title('Daily Precipitation - Olpe vs HSK')
ax.set_ylabel('Daily Precipitation (mm)')
#ax.set_xlabel('Date')
plt.xticks(rotation=45)
plt.setp(ax, ylim=custom_ylim)
plt.legend()
plt.savefig('Olpe_vs_HSK_bar_1.png')
plt.show()



labels = df_day_may.DATE
x = np.arange(len(labels))
width = 0.5
fig, ax = plt.subplots(figsize=(20,7))
ax.bar(x-width/2, df_day_may.R_mean_olpe, width, label = 'Olpe')
ax.bar(x+width/2, df_day_may.R_mean_hsk, width, label = 'HSK')

ax.set_xticks(x)
ax.set_xticklabels(map(line_format, df_day_may.index))

ax.set_title('Daily Precipitation - Olpe vs HSK')
ax.set_ylabel('Daily Precipitation (mm)')
ax.set_xlabel('Date')
plt.xticks(rotation=45)
plt.setp(ax, ylim=custom_ylim)
plt.legend()
plt.savefig('Olpe_vs_HSK_bar_2.png')
plt.show()



labels = df_day_june.DATE
x = np.arange(len(labels))
width = 0.5
fig, ax = plt.subplots(figsize=(20,7))
ax.bar(x-width/2, df_day_june.R_mean_olpe, width, label = 'Olpe')
ax.bar(x+width/2, df_day_june.R_mean_hsk, width, label = 'HSK')

ax.set_xticks(x)
ax.set_xticklabels(map(line_format, df_day_june.index))

ax.set_title('Daily Precipitation - Olpe vs HSK')
ax.set_ylabel('Daily Precipitation (mm)')
ax.set_xlabel('Date')
plt.xticks(rotation=45)
plt.setp(ax, ylim=custom_ylim)
plt.legend()
plt.savefig('Olpe_vs_HSK_bar_3.png')
plt.show()



labels = df_day_july.DATE
x = np.arange(len(labels))
width = 0.5
fig, ax = plt.subplots(figsize=(20,7))
ax.bar(x-width/2, df_day_july.R_mean_olpe, width, label = 'Olpe')
ax.bar(x+width/2, df_day_july.R_mean_hsk, width, label = 'HSK')

ax.set_xticks(x)
ax.set_xticklabels(map(line_format, df_day_july.index))

ax.set_title('Daily Precipitation - Olpe vs HSK')
ax.set_ylabel('Daily Precipitation (mm)')
ax.set_xlabel('Date')
plt.xticks(rotation=45)
plt.setp(ax, ylim=custom_ylim)
plt.legend()
plt.savefig('Olpe_vs_HSK_bar_4.png')
plt.show()
```
